```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15,10)

pd.options.mode.chained_assignment = None
```

```python
# Read in the data

df = pd.read_csv(r'C:\Users\Desktop\Desktop\Data Analytics\sample_data\Iracing\JandTdata.csv')
```

```python
# Let's look at the data
df.head ()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Let's take a look at the traction circle for both drivers
sns.relplot (x="LatAccel_G", y="LongAccel_G", hue="Driver", data = df)
```

```python
# Let's see just the traction cirlce for Josh
sns.relplot (x="LatAccel_G", y="LongAccel_G", size= "Speed_ft/s", data = df[df["Driver"]=="Josh"])
```

```python
# Let's take a look at Travis's traction cirlce
sns.relplot (x="LatAccel_G", y="LongAccel_G", size= "Speed_ft/s", data = df[df["Driver"]=="Travis"])
```

```python
# Graph the speed over distance between the two
sns.lineplot (data=df, x="LapDist_ft", y="Speed_ft/s", hue="Driver")
plt.figure(figsize = (20,20))
```

```python
#show the difference in throttle application
sns.lineplot (data=df, x="LapDist_ft", y="Throttle_%", hue="Driver")
plt.gcf().set_size_inches(20,5)
```

```python
#show the difference in brake application
sns.lineplot (data=df, x="LapDist_ft", y="Brake_%", hue="Driver")
plt.gcf().set_size_inches(20,5)
```

```python
#Show the lap time over distance
sns.lineplot (data=df, x="LapDist_ft", y="LapCurrentLapTime_s", hue="Driver")
plt.gcf().set_size_inches(20,5)
```
